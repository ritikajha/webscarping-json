```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/openings/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
openings=html.find_all("div")

    
f=open("openings_content_text",'w',encoding='UTF-8')
for i in range(0,len(openings)):
    #f.write(i['class']+'\n')
    f.write(openings[i].get_text())
f.close()
```

```python
with open("openings_content_text","r",encoding="UTF-8") as f:
    all_openings=f.read()
    print(all_openings)
```

```python
with open("mopenings_content.json","w") as f:
    json.dump(all_openings, f)
```
