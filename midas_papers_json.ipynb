```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/papers/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
papers=html.find_all("div")

    
f=open("papers_content_text",'w',encoding='UTF-8')
for i in range(0,len(papers)):
    #f.write(i['class']+'\n')
    f.write(papers[i].get_text())
f.close()
```

```python
with open("papers_content_text","r",encoding="UTF-8") as f:
    all_papers=f.read()
    print(all_papers)
```

```python
with open("mpapers_content.json","w") as f:
    json.dump(all_papers,f)
  
```
