```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/projects/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
content=html.find_all("div")

    
f=open("projects_content_text",'w',encoding='UTF-8')
for i in range(0,len(content)):
    #f.write(i['class']+'\n')
    f.write(content[i].get_text())
f.close()
```

```python
with open("projects_content_text","r",encoding="UTF-8") as f:
    all_text=f.read()
    print(all_text)
```

```python
with open("mprojects_content.json","w") as f:
    json.dump(all_text, f)
```
