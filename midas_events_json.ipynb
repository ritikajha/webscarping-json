```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/events/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
events=html.find_all("div")

    
f=open("events_content_text",'w',encoding='UTF-8')
for i in range(0,len(events)):
    #f.write(i['class']+'\n')
    f.write(events[i].get_text())
f.close()
```

```python
with open("events_content_text","r",encoding="UTF-8") as f:
    all_events=f.read()
    print(all_events)
```

```python
with open("mevents_content.json","w") as f:
    json.dump(all_events, f)
```
