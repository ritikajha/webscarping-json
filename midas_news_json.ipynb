```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/news/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
news=html.find_all("div")

    
f=open("news_content_text",'w',encoding='UTF-8')
for i in range(0,len(news)):
    #f.write(i['class']+'\n')
    f.write(news[i].get_text())
f.close()
```

```python
with open("news_content_text","r",encoding="UTF-8") as f:
    all_news=f.read()
    print(all_news)
```

```python
with open("mnews_content.json","w") as f:
    json.dump(all_news, f)
```
