```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/blog/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
blog=html.find_all("div")

    
f=open("blog_content_text",'w',encoding='UTF-8')
for i in range(0,len(blog)):
    #f.write(i['class']+'\n')
    f.write(blog[i].get_text())
f.close()
```

```python
with open("blog_content_text","r",encoding="UTF-8") as f:
    all_blog=f.read()
    print(all_blog)
```

```python
with open("mblog_content.json","w") as f:
    json.dump(all_blog, f)
```
