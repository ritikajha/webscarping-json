```python
import json
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
midas="http://midas.iiitd.edu.in/team/"
page=urllib.request.urlopen(midas)
html=BeautifulSoup(page)
```

```python
name=html.find_all("div")

    
f=open("team_content_text",'w',encoding='UTF-8')
for i in range(0,len(name)):
    #f.write(i['class']+'\n')
    f.write(name[i].get_text())
f.close()
```

```python
with open("team_content_text","r",encoding="UTF-8") as f:
    all_name=f.read()
    print(all_name)
```

```python
with open("mteam_content.json","w") as f:
    json.dump(all_name, f)
```
